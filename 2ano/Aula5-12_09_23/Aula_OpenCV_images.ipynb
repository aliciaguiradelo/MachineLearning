*Alterando o tamanho de uma imagem*

O redimensionamento da imagem pode ser feito na OpenCV através do comando:

cv2.resize ( imagem, tamanho, interpolação )
os parâmetros imagem e tamanho da função acima são obrigatórios.

https://www.tutorialkart.com/opencv/python/opencv-python-resize-image/#gsc.tab=0

O tamanho é dado por uma tupla (W,H), onde W é a largura (número de colunas) e H é a altura (número de linhas)

Quanto mais pixels estiverem concentrados num espaço, maior vai ser a qualidade e nitidez dessa imagem.

Os softwares de edição de imagem criam ou reduzem pixels na imagem a partir dos pixels existentes.

O processo de “criação” chama-se interpolação.



```python
imagem = cv2.imread("NATUREZA_1.png")
```

```python
plt.imshow(imagem)
# plt.axis('off')
plt.show()
```

```python
imagem.shape
```

```python
#criando uma nova variável (imagem2), pego a imagem anterior e aumento o tamanho dele utiliizando o método RESIZE
imagem2 = cv2.resize(imagem, (600, 400) )
```

```python
imagem2.shape
```

```python
plt.imshow(imagem2)
plt.axis('off')
plt.show()
```

```python
# PARTE 05
# Repita o item anterior, agora testando e comparando as flags de interpolação Vizinhos Mais Próximos, Bilinear e Bicúbica.
# Redimensionando para 10x maiores com interpolação Vizinhos Mais Próximos
image_nn = cv2.resize(image, None, fx=10, fy=10, interpolation=cv2.INTER_NEAREST)
plt.imshow(cv2.cvtColor(image_nn, cv2.COLOR_BGR2RGB))
plt.title("Vizinhos Mais Próximos")
plt.show()

# Redimensionando para 10x maiores com interpolação Bilinear
image_bl = cv2.resize(image, (image.shape[1] * 10, image.shape[0] * 10), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(image_bl, cv2.COLOR_BGR2RGB))
plt.title("Bilinear")
plt.show()

# Redimensionando para 10x maiores com interpolação Bicúbica
image_bc = cv2.resize(image, (image.shape[1] * 10, image.shape[0] * 10), interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(image_bc, cv2.COLOR_BGR2RGB))
plt.title("Bicúbica")
plt.show()
```
