```python
# Importando as bibliotecas OpenCV, DLib e im_show
import cv2
import dlib
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread('people2.jpg')
```

```python
detector_face_cnn = dlib.cnn_face_detection_model_v1('/content/mmod_human_face_detector.dat')
# Quando executamos deu o erro "Error while calling cudaGetDevice". Está relacionado com cuda cores. Que se refere a GPU que são hardwares incorporados.
# Para resolver vá no menu acima clique em Ambiente de excecução -> alterar tipo de ambiente de execução -> seleciona T4 GPU
# Depois vali ter que importar os arquivos denovo porque reiniciou o ambiente e vai ter que executar desde a primeira célula
```

```python
deteccoes = detector_face_cnn(imagem, 1)
for face in deteccoes:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(l,t,r,b,c)
  cv2.rectangle(imagem,(l,t), (r,b), (255, 255, 0), 1)
cv2_imshow(imagem)
```
