#### Histogram Of Oriented Gradients (HOG)
o elgorítmo hog não dá falso positivo e tem uma precisão muito maior (detectou o rosto inclinado e não pegou o joelho como se fosse um rosto)
https://learnopencv.com/histogram-of-oriented-gradients/

```python
# Importando as bibliotecas OpenCV, DLib e im_show
# Usamos a biblioteca do open cv para pegar as imagens
# Biblioteca mais aprimorada de reconhecimento de faces
# A biblioteca im_show é usada para plotar a matriz da imagem, em outras palavras, para mostrar a imagem na tela
import cv2
import dlib
from google.colab.patches import cv2_imshow
```

```python
# Insira aqui o seu arquivo
imagem = cv2.imread('people2.jpg')
```

```python
# O método .get_frontal_face_detector() faz a detecção de faces automaticamente sem modelo pré-treinado.
# É um método da dlib
detector_face_hog = dlib.get_frontal_face_detector()
```

```python
deteccoes = detector_face_hog(imagem, 2) # O segundo parâmetro é semelhante ao Scale Factor do Opencv (Haarcascades)
```

```python
deteccoes
```

```python
# Conta quantas faces ele detecta no total
len(deteccoes)
```

```python
for face in deteccoes:
  print(face.left())
  print(face.top())
  print(face.right())
  print(face.bottom())
```

```python
for face in deteccoes:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagem, (l, t), (r, b), (0, 255, 0), 1)
cv2_imshow(imagem)
```

```python
imagem = cv2.imread('people2.jpg')
detector_face_hog = dlib.get_frontal_face_detector()
deteccoes = detector_face_hog(imagem, 2)
for face in deteccoes:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagem, (l, t), (r, b), (0, 255, 0), 1)
cv2_imshow(imagem)
```
