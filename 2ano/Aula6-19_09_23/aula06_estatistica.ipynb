```python
# Vamos acessar os registros a partir do rótulo escolhido
df[['price']]
```

```python
# Média da população
media = df[['price']].mean()
media
```

```python
# Mediana da população
mediana = df[['price']].median()
mediana
```

```python
# Desvio padrão e variância (é uma medida de dispersão)
# Calcula a distancia do datapoint (valor do dado) e a reta da média
desvio_padrao = df[['price']].std()
desvio_padrao
```

Geração de amostras

```python
# Primeira amostra
# No iloc, eu passo uma lista, seleciono as linhas e colunas que eu quero
a1 = df.iloc[0:3000,:]
# O : serve para dizer que são todos
a1
```

```python
# Calculando média para a amostra a1
mediaA1 = a1[['price']].mean()
mediaA1
```

```python
# Calculando mediana para a amostra a1
medianaA1 = a1[['price']].median()
medianaA1
```

```python
# Calculando desvio para a amostra a1
desvioA1 = a1[['price']].std()
desvioA1
```

```python
# Segunda amostra
# No iloc, eu passo uma lista, seleciono as linhas e colunas que eu quero
a2 = df.iloc[3000:6000,:]
# O : serve para dizer que são todos
a2
```

```python
# Calculando média para a amostra a2
a2[['price']].mean()
```

```python
# Calculando mediana para a amostra a2
a2[['price']].median()
```

```python
# Calculando desvio para a amostra a2
a2[['price']].std()
```
